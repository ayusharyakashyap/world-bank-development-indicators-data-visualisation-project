```python
# Load the dataset
file_path = '/Users/ayusharyakashyap/Desktop/Semester 5/Data Visualization/A1 Project/world_bank_development_indicators.csv'
df = pd.read_csv(file_path)
```

```python
# Convert 'date' column to datetime if it's not already in datetime format
df['date'] = pd.to_datetime(df['date'], errors='coerce')

# Extract the year from the 'date' column
df['year'] = df['date'].dt.year

# Filter the DataFrame for the years 2008 and 2021
df_filtered = df[df['year'].isin([2010, 2021])]

# Display the filtered data
print(df_filtered)

# Optional: If you want to save the filtered data to a CSV
df_filtered.to_csv('/Users/ayusharyakashyap/Desktop/Semester 5/Data Visualization/A1 Project/data_2010.csv', index=False)
```

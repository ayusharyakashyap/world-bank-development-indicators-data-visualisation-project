```python
import pandas as pd
import numpy as np
import os 
```

```python
# Load the dataset
file_path = '/Users/ayusharyakashyap/Desktop/Semester 5/Data Visualization/A1 Project/world_bank_development_indicators.csv'
df = pd.read_csv(file_path)
```

```python
# Convert the 'Year' column to datetime format if necessary
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')

# Extract rows with the country name "Ukraine" and from the year 1989 onwards
ukraine_data = df[(df['country'].str.startswith('Ukraine')) & (df['date'].dt.year >= 1989)]

# Save the new dataset to a CSV file
ukraine_data.to_csv('/Users/ayusharyakashyap/Desktop/Semester 5/Data Visualization/A1 Project/Ukraine_data.csv', index=False)

# Display the first few rows of the new dataset
print(ukraine_data)
```

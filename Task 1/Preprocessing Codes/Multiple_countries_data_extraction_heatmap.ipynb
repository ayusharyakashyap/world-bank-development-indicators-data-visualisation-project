```python
import pandas as pd
import numpy as np
import os 
```

```python
# Load the dataset
file_path = '/Users/ayusharyakashyap/Desktop/Semester 5/Data Visualization/A1 Project/Filtered CSV/world_bank_development_indicators_with_gdp_per_capita.csv'
df = pd.read_csv(file_path)
```

```python
# Convert the 'Year' column to datetime format if necessary
df['date'] = pd.to_datetime(df['date'], format='%Y-%m-%d')

# List of country prefixes to filter
countries = ('Brunei Darussalam', 'Cambodia', 'Indonesia', 'Lao PDR', 'Malaysia', 'Myanmar', 'Philippines', 'Singapore', 'Thailand', 'Vietnam', 'East Timor')

# Extract rows with the specified country names and from the year 2000 onwards
selected_countries_data = df[(df['country'].str.startswith(countries)) & (df['date'].dt.year >= 2000)]

# Save the new dataset to a CSV file
selected_countries_data.to_csv('/Users/ayusharyakashyap/Desktop/Semester 5/Data Visualization/A1 Project/selected_countries_2000+_heatmap_data.csv', index=False)

# Display the first few rows of the new dataset
print(selected_countries_data)
```
